```python
df=df.drop("country",axis=1)
```

```python
scaler = StandardScaler()
ndf=scaler.fit_transform(df)
```

```python
f_df=pd.DataFrame(ndf,columns=df.columns)
f_df
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.title("dendrogram")
shc.dendrogram(shc.linkage(f_df,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
d=cluster.fit_predict(f_df)
d
```

```python
D=pd.DataFrame(d)
v=D[0].value_counts()
```

```python
plt.pie(v,explode=[0.02,0.02],autopct="%1.1f%%",shadow=True)
plt.title("Cluster Percentage in Pie")
plt.show()
```

```python
import seaborn as sns
sns.countplot(x=0,data=D)
plt.xlabel("clusters")
plt.ylabel("cluster data counts")
plt.title("Cluster Counts")
plt.show()
```

### Normalizing Data

```python
from sklearn.preprocessing import normalize
```

```python
nrml_data=normalize(df)
nrml_data=pd.DataFrame(nrml_data,columns=df.columns)
nrml_data.head()
```

### plotting Dentrogram

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.title("dendrogram")
shc.dendrogram(shc.linkage(nrml_data,method='ward'))
plt.show()
```

### Plotting threshold line

```python
shc.dendrogram(shc.linkage(nrml_data,method="ward"))
plt.axhline(y=6,color="r",linestyle="--")
plt.show()
```

### Hierarchical Clustering model

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
d=cluster.fit_predict(nrml_data)
d
```

### Vizualization

```python
plt.figure(figsize=(6,6))
plt.scatter(nrml_data["Milk"],nrml_data["Grocery"],c=cluster.labels_)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(nrml_data["Fresh"], nrml_data["Milk"],c=cluster.labels_)
```
